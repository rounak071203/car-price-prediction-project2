```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['year'].unique()
```

```python
df['year'] = pd.to_numeric(df['year'], errors = 'coerce')
```

```python
df.dropna(subset = ['year'], inplace = True)
```

```python
df['year'] = df['year'].astype(int)
df.info()
```

```python
df['Price'].unique()
```

```python
for i in df.index:
    price = df.loc[i, 'Price']
    price = price.replace(',', '')
    df.loc[i, 'Price'] = price
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors = 'coerce')
```

```python
df.dropna(subset = ['Price'], inplace = True)
```

```python
for i in df.index:
    kms_driven = df.loc[i, 'kms_driven']
    kms_driven = kms_driven.replace(',', '')
    kms_driven = kms_driven.replace(' kms', '')
    df.loc[i, 'kms_driven'] = kms_driven
```

```python
df['kms_driven'] = pd.to_numeric(df['kms_driven'], errors = 'coerce')
```

```python
df.dropna(subset = ['kms_driven'], inplace = True)
```

```python
df['fuel_type'].unique()
```

```python
df.dropna(subset = ['fuel_type'], inplace = True)
```

```python
df.info()
```

```python
df['company'].unique()
```

```python
len(df['name'].unique())
```

```python
for i in df.index:
    name = df.loc[i, 'name']
    company = df.loc[i, 'company']
    name = name.replace(company, '').strip()
    name = " ".join(name.split()[:2])
    df.loc[i, 'name'] = name
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df.to_csv('cleaned_data.csv', index = False)
```
